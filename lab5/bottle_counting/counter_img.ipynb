```python
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator
from matplotlib import pyplot as plt


# load model
model = YOLO("yolov8n.pt")
model.fuse()

img = cv2.imread("dataset/exemple.png")


results = model(img)  
result = results[0]

print(result.names)

res_plotted = result.plot()

plt.imshow(res_plotted)
plt.title('res_plotted')
plt.show()

color = (0, 0, 255)

# count the bottles in result.boxes.data 

count_bottles = 0
for label in result.boxes.data:

    if result.names[label[-1].item()] != 'bottle':
        continue

    count_bottles += 1

print("Number of bottles : ", count_bottles)
```
