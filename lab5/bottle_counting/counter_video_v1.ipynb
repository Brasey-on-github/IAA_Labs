```python
import cv2
from ultralytics import YOLO
from matplotlib import pyplot as plt

# Load model
model = YOLO("yolov8n.pt")
model.fuse()

# Initialize video capture (0 for webcam or provide a path to a video file)
cap = cv2.VideoCapture(0)

total_bottles = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform detection on the frame
    results = model(frame)
    result = results[0]

    # Initialize bottle count for this frame
    count_bottles = 0

    # Count bottles in the current frame
    for label in result.boxes.data:
        if result.names[label[-1].item()] == 'bottle':
            count_bottles += 1

    total_bottles += count_bottles

    # Annotate frame with the detection results
    res_plotted = result.plot()

    # Display the frame with annotations
    cv2.imshow('YOLO Detection', res_plotted)

    # Break loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()

print("Total number of bottles detected:", total_bottles)
```
